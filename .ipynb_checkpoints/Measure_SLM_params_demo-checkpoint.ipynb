```python
def max_researcher(signal, step, start):
    start = 0
    max_signals = []
    len_s = len(signal)
    while start + step < len_s:
        info_signal = signal[start: start + step]
        max_signal = np.max(info_signal) 
        max_signals.append(max_signal)
        start += step
    max_signals = np.array( max_signals)
    
    return max_signals 
            
    
```

```python
def unpack(way, start_point_1 , start_point_2, start_point_3, min_frame, max_frame):
    start_time = t.time()
    first_point_max = []
    second_point_max = []
    third_point_max = []
    min_count = []
    for i in range(min_frame,max_frame):
        num = digits(i)
        target = (5-num)*"0"+str(i)
        newway = way + '{}.tif'.format(target)
        img = np.array(Image.open(newway)) 
        minmin_c = np.min(img)
        min_count.append(min_c)
    mean_min = np.mean(np.array(min_c))
    print(mean_min)
    for i in range(min_frame,max_frame):
        num = digits(i)
        target = (5-num)*"0"+str(i)
        newway = way + '{}.tif'.format(target)
        img = np.array(Image.open(newway))
        img = img - mean_min
        first = np.sum(img[start_point_1[0]:start_point_1[0]+window_size, start_point_1[1]:start_point_1[1]+window_size])
        second = np.sum(img[start_point_2[0]:start_point_2[0]+window_size,start_point_2[1]:start_point_2[1]+window_size])
        third = np.sum(img[start_point_3[0]:start_point_3[0]+window_size,start_point_3[1]:start_point_3[1]+window_size])

        first_point_max.append(first)
        second_point_max.append(second)
        third_point_max.append(third)
        
        
    first_point = np.array(first_point_max)
    second_point = np.array(second_point_max)
    third_point = np.array(third_point_max)
    print("--- %s seconds ---" % (t.time() - start_time))
    return first_point, second_point, third_point
    
```

```python
def par(par_sign):
    for i in range(len(par_sign)):
        front_min = par_sign[i][0]
        front_max = par_sign[i][1]
        front_time = front_max - front_min
        
        back_max = par_sign[i][2]
        back_min = par_sign[i][3]
        back_time = back_min - back_max
        
        width = back_max - front_max 
        if i == 0:
            full_front_time  = front_time
            full_back_time =  back_time
            full_width  = width
        else:
            full_front_time = np.hstack([full_front_time, front_time])
            full_back_time = np.hstack([full_back_time, back_time])
            full_width = np.hstack([full_width, width])
            
    back_max_1 = par_sign[0][2]
    back_max_2 = par_sign[1][2]
    back_max_3 = par_sign[2][2]
    min_len = min([len(back_max_1), len(back_max_2), len(back_max_3)])
    back_max_1 = back_max_1[: min_len]
    back_max_2 = back_max_2[:min_len]
    back_max_3 = back_max_3[:min_len]
    period_3 = []
    for i in range (len(back_max_1)-1):
        period = back_max_1[i+1] - back_max_3[i]
        period_3.append(period)
    period_3 = np.array(period_3)
    
    full_Period = np.hstack([back_max_2 -back_max_1, back_max_3 - back_max_2,period_3  ])
    print("Количество передних фронтов = ", len(full_front_time))
    print("Количество задних фронтов = ", len(full_back_time))
    print("Количество длительностей импульсов = ", len(full_width))
    print("Количество  периодов  = ", len(full_Period))       
        
    return [full_front_time,full_back_time, full_width, full_Period]
        
```
